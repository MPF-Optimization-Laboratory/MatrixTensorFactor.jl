# How to Use the `MultiFactorize` function

Welcome to this tutorial on how to use the `MultiFactorize` function from the `BlockTensorDecomposition` package in Julia! In this tutorial, we'll walk you through the understanding and practical use of the function in different scenarios.

---

## Overview of `MultiFactorize`

`MultiFactorize` is a function in the `BlockTensorDecomposition` package that allows you to perform a Tucker-1 decomposition faster when your data is continuous. 

- **Purpose**: Speed up a Tucker-1 decomposition
- **Input(s)**: A 3 Dimensional tensor that is continuous accross the third dimension (code to further be generalized to any dimension)
- **Output(s)**: A core tensor and a proportion matrix

---
## Understanding how `MultiFactorize` works

This function works by performing the `factorize` function repeadidly on a smaller tensor than the original and using the result as a warm start to the next iteration. 

Suppose we define our original tensor as: 
$
Y = \left[ Y_{ijk} \right], \quad i \in \{1, 2, \dots, I\}, \; j \in \{1, 2, \dots, J\}, \; k \in \{1, 2, \dots, K\}.
$

The algorithm will start by sampling the first, middle and end point of $Y$. Consider this new sampled tensor to be $Y_1$. 

Then:
$Y_1 = \left[ Y_{ijk} \right], \quad i \in \{1, 2, \dots, I\}, \; j \in \{1, 2, \dots, J\}, \; k \in \{1, \frac{K}{2}, K\}.$

$Y_1$ then gets passed into the `factorize` function where we perform alternating block gradient descent on randomly initilized factors. Call these randomly initialized factors $A_1$ and $B_1$ where $A_1$ is the proportion matrix and $B_1$ is the core tensor. 

Since we are looking for the result of the factorization on $Y$ and not $Y_1$, we linearly interpolate $B_1$ to the next size. This next size is dynamically determined by how many new points we are adding in our next call to factorize. For example, the next iteration would linearly interpolate 2 new points. The one between the first slice and the middle sclice, and the one between the middle slice and the end slice. We can call this new core tensor $B_2$. It is also important to note that since $A_1$ is of size $I \times R$, we do not need to change anything about $A_1$ for the next iteration and it simply gets refined after every iteration. 

In the second iteration we will sample $Y$ to get $Y_2$ by defining:
$Y_2 = \left[ Y_{ijk} \right], \quad i \in \{1, 2, \dots, I\}, \; j \in \{1, 2, \dots, J\}, \; k \in \{1, \frac{K+2}{4}, K/2, \frac{3K}{4} K\}.$

We will then use this sampled tensor $Y_2$ to pass into the `factorize` function. The difference is that this time we do not have randomly initilized factors. We have $A_1$ and $B_2$ which can give a much better initilization and save a lot of computational power. 

This algorithm repeats until our sampled tensor $Y_N$ becomes $Y$. Then the algorithm ends by performing the factorization on $Y$ with the factors $A_1$ and $B_{N}$ as a warm start to the factorization.

## Basic Usage

`MultiFactorize` is intended to be a substitute function for `factorize` under specific use cases. Let us first explore what happens in the unintended case. 


### Unintended use
Let us create a random tensor to be decomposed by the Tucker-1 decomposition. 



```julia
using BlockTensorDecomposition
factorize = BlockTensorDecomposition.factorize
MultiFactorize = BlockTensorDecomposition.MultiFactorize;

Y = abs_randn(30, 50, 60);
```

This creates a tensor of dimensions $30$x$50$x$60$. We will now attempt to decompose it with the intended `factorize` function and compare it to the `MultiFactorize` function.

Let us first run the two functions with a small input so they can compile. 

```julia
test = abs_randn(3, 3, 3);
factorize(test, 
    rank=2, 
    tolerence=(0.35), 
    converged=(RelativeError),
    constraints=[nonnegative!, nonnegative!]);
MultiFactorize(test, 
    rank=2,
    tolerence=(0.35), 
    converged=(RelativeError),
    constraints=[nonnegative!, nonnegative!]);
```

Now let us try it on the larger tensor.

```julia
@time factorize(Y; 
    rank=20,
    tolerence=(0.35),
    converged=(RelativeError),
    constraints=[nonnegative!, nonnegative!]);
```

```julia
@time MultiFactorize(Y; 
    rank=20,
    tolerence=(0.35),
    converged=(RelativeError),
    constraints=[nonnegative!, nonnegative!]);
```

We notice that there is not a significant difference in the time it takes to run `factorize` vs `MultiFactorize`. This is because in `MultiFactorize`, the previous scale does not give a good initialization to the next. This is because we linearly interpolated half of the points in the current 'warm start', but in reality the interpolation does not make sense since every point is random. 

### Intended use

Now we will explore the power of `MultiFactorize`. First to do this we must understand how to organize our data. For the best use case we organize our data in a 3D tensor as such:

Where every fiber accross the third dimension sums to 1. We also consider the fact that each slice of the tensor is a convex combination of a set of linearly independent slices. This looks like the following:

These two assumptions are valid towards real life data. To see a real-life geological application, check out the paper listed at https://friedlander.io/publications/2024-sediment-source-analysis/. 

We will create some synthetic data with these properties using the `generate_tensor_streams` function. To learn more about the `generate_tensor_streams` function, consult its docstring. 

```julia
generate_tensor = BlockTensorDecomposition.generate_tensor_streams;

rank = 10
tensor_size = (20, 10, 200)
Y, matrix, Y_prime = generate_tensor(tensor_size, rank);
```

We will now do the same test comparing the `factorize` function time to the `Multifactorize` function time. This time we use the simplex constraints in order to get the practical application described by the paper. 


In our call to `factorize`, we set the maximum iterations to infinity so we do not stop early before the convergence criteria is met. We do not set a maximum iterations for the multi-factorize function since we want each interal iteration to stop at some point since often the internal iterations do not fully converge. 

```julia
@time factorize(Y;
    rank=rank,
    momentum=true,
    tolerence=(0.05),
    converged=(RelativeError),
    constrain_init=true,
	constraints = [simplex_12slices!, simplex_rows!],
    maxiter=Inf);
```

```julia
@time MultiFactorize(Y;
    rank=rank,
    momentum=true,
    tolerence=(0.05),
    converged=(RelativeError),
    constrain_init=true,
    constraints=[simplex_12slices!, simplex_rows!]);
```

From this we can see a great difference in time between the two functions. This is no coincedence. As the tensor of these conditions grows in size, the better it becomes to use `MultiFactorize`.

## Discussions 

### Trying to make it even faster
One of the biggest issues with this current work is deciding when to stop the internal iterations and move on to the next scale. This is an important question since if we choose the internal iterations to have the same criteria for convergence as the final specified criteria, it can lead to a lot of unecessary work done. For example, if we have a very strict RelativeError but a very large maxiter, then it would not make sense to use this criteria on $Y_1$ since it may even be impossible to converge with RelativeError alone and thus will perform every maxiter without making any meaningful progress. This brings up the question of how can we add other criteria for convergence that we can potentially hit first and move on to the next iteration. 

One approach that was tested was to add a new convercence criteria (along with the old ones) that hits if we do not see any meaningful progress to the RelativeError in our block update. This criteria was called ObjectiveRatio. This is not a terrible approach but was two major setbacks. The first problem is that ObjectiveRatio requires a an input parameter that needs to be tuned. This parameter describes the minimum improvement accross block updates in order to continue. For example if we set the parameter to be 0.001 we would say that if a block update does not improve in RelativeError by at least 0.1%, then we say that we have converged and move to the next scale. This method appears to perform really well in certain cases but also has certain drawbacks. First, what should the parameter be? In order for this to be an effective approach, it appears as if the parameter is dependent on both the original tensor size as well as the desired final RelativeError. Exploring how the tensor size and RelativeError impact the effectiveness of the ObjectiveRatio would be an interesting avenue for further research. 

Another approach to speed up this function was to deterministically change the maxiter parameter on each scale. For example this looked like using 2 alternating block gradient descent steps (ABGDS) on $Y_1$, 4 AGBDS on $Y_2$, 8 ABGDS on $Y_3$, ..., (original input) maxiter ABGDS on $Y$. This logscale approach to determine the internal maxiter was the most promising since it kept the total number of possible iterations in a linear asymptotic bound of $O(\text{maxiter})$, which would make it easier for the user to interpret how maxiter influences the function. This approach was also tried with other deterministic methods such as linear growth in the parameter, but nothing jumped out as being a clear cut best method. Overall, this approach still runs much faster than the factorize function but is still unclear on when exactly it should be used. 

Due to the fact that for all constant ObjectiveRatio parameters and deterministically changing maxiter parameters, one could find a counterexample where these strategies actually made the function slower, it was decided not to include them in the final code. The methods are still present in the file OtherMultiFactorize.jl if time permits these methods to be further examined in the future. 

### `MultiFactorize` vs `Factorize` with random data

It is interesting to note that MultiFactorize seems to perform notably better than Factorize
even when the data is random. This is an unexpected result as we would expect that MultiFactorize to perform slower since the linear interpolation doesn't actually help the next iteration since the data is random and not continuous. The current hypothesis is that MultiFactorize is faster since it takes less gradient steps to reach the desired RelativeError in the final iteration. Since calculating the gradient is the most costly step in either function, MultiFactorize can still perform better since the gradient steps on the smaller scales are less costly than a gradient step on a large scale. So even with the interpolation not being benificial, it still doesn't take many gradient steps on each scale to reach the desired RelativeError since more than half of the points are alredy tuned to that error. Getting a more conclusive and detailed explanation as to why MultiFactorize is running faster is another possible area to continue research. 


## Future work
In addition to the future work described in the discussion, there is even more to be done to make this project even better as listed below. 
### Generalizing to further dimensions and decopositions
This MultiFactorize approach has only been implemented for the Tucker1 decomposition where the input tensor is of 3 dimmensions and only one of the dimensions is continuous. The next step would be to generalize this to a N-dimensional tensor where we can also have <1 dimensions that are continuous. From here we would then want to generalize to the general tucker decomposition and not just tucker1. After this explore further decompositions such as the (CP-Decomposition) and see if this multifactorize method still applies, or if there are any other techniques we can do to speed up the decomposition. The goal would be to generalize as much of the code as possible. 
### Memory management
There are some areas of the code where there can be more care taken in regards to memory management. For example when we sample each tensor $Y_i$, we create a new tensor all together. This was done because there are a lot of operations being done on this tensor and so it makes sense to create a new one, but we do not know for sure. It is also possible to create a new tensor with pointers to the relevant parts of the original which is likely a worse option to chose but worth exploring nonetheless. 
### CUDA Programming
In today's world, it seems very relevant to make this code GPU ready. To do this we could use the CUDA framework which could accelerate computationally heavy tasks. It would be interesting to explore what is possible in CUDA programming in this project as it seems that large tensors may not be able to entirly fit onto the GPU's on chip memory. Regardless, parallelizing as much of the code as possilbe will be extremely benificial to this project.
